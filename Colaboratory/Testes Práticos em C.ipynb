##**Teste 1: Alocação de Memória**

**Objetivo**: Validar o espaço físico (bytes) e o endereçamento lógico de variáveis primitivas.

```python
# <<< Crie o arquivo aqui antês de rodar o comando! <<<

%%writefile memoria.c
#include <stdio.h>
#include <locale.h>

// Definição de cores para o terminal
#define RESET  "\033[0m"
#define VERDE  "\033[1;32m"
#define CIANO  "\033[1;36m"
#define LARANJA "\033[38;5;215m"

int main() {
    setlocale(LC_ALL, "Portuguese");

    // Identificação com cor Verde
    printf(VERDE "==========================================\n");
    printf("  RELATÓRIO TÉCNICO: ANÁLISE DE MEMÓRIA   \n");
    printf("  Acadêmico: Murilo Guimarães             \n");
    printf("==========================================\n\n" RESET);

    // Tabela com cor Ciano
    printf(CIANO "TABELA DE ALOCACAO (sizeof):\n");
    printf("------------------------------------------\n" RESET);
    printf("Tipo CHAR:    " LARANJA "%zu byte\n" RESET, sizeof(char));
    printf("Tipo INT:     " LARANJA "%zu bytes\n" RESET, sizeof(int));
    printf("Tipo FLOAT:   " LARANJA "%zu bytes\n" RESET, sizeof(float));
    printf("Tipo DOUBLE:  " LARANJA "%zu bytes\n" RESET, sizeof(double));
    printf(CIANO "------------------------------------------\n\n" RESET);

    // Exemplo de endereço com cor Laranja
    int variavelExemplo = 10;
    printf(VERDE "EXEMPLO DE ENDEREÇAMENTO:\n" RESET);
    printf("Valor da variável: " LARANJA "%d\n" RESET, variavelExemplo);
    printf("Endereço físico:   " LARANJA "%p\n" RESET, (void*)&variavelExemplo);
    printf(VERDE "==========================================\n" RESET);

    return 0;
}
```

*Execute com:*

```python
# Compila o arquivo criado, gera um executável e executa o programa
!gcc memoria.c -o memoria && ./memoria
```

##**Teste 2: Precisão de Dados (<code>float</code> vs <code>double</code>)**

**Objetivo:** Demonstrar por que o <code>double</code> ocupa o dobro de espaço: ele armazena muito mais informação decimal.

```python
# <<< Crie o arquivo aqui antês de rodar o comando! <<<

%%writefile precisao.c
#include <stdio.h>

#define RESET    "\033[0m"
#define VERDE    "\033[1;32m"
#define LARANJA  "\033[38;5;215m"

int main() {
    float f = 1.123456789f;
    double d = 1.123456789123456789;

    printf(VERDE "--- TESTE DE PRECISÃO NUMÉRICA ---\n\n" RESET);

    printf("Valor original: 1.123456789123456789\n\n");

    printf("Float (4 bytes):  " LARANJA "%.10f\n" RESET, f);
    printf("Double (8 bytes): " LARANJA "%.20f\n" RESET, d);

    printf(VERDE "\nNota: Perceba como o Float 'arredonda' antes do Double.\n" RESET);
    return 0;
}
```

*Execute com:*

```python
# Compila o arquivo criado, gera um executável e executa o programa
!gcc precisao.c -o precisao && ./precisao
```

##**Teste 3: Endereçamento Dinâmico (A Prova da RAM)**


**Objetivo:** Este teste serve para mostrar que a memória é viva. Ao rodar duas vezes, os endereços mudam.

```python
# <<< Crie o arquivo aqui antês de rodar o comando! <<<

%%writefile dinamico.c
#include <stdio.h>

#define RESET    "\033[0m"
#define LARANJA  "\033[38;5;215m"

int main() {
    int a = 5, b = 10;

    printf("Endereço de A: " LARANJA "%p\n" RESET, (void*)&a);
    printf("Endereço de B: " LARANJA "%p\n" RESET, (void*)&b);

    return 0;
}
```

*Execute com:*

```python
# Compila o arquivo criado, gera um executável e executa o programa
!gcc dinamico.c -o dinamico && ./dinamico
```

##**Teste 4: Overflow de memória**


**Objetivos:** Analisar fisicamente a alocação de bytes, o endereçamento dinâmico na RAM e o fenômeno de overflow em variáveis da linguagem C.

```python
# <<< Crie o arquivo aqui antês de rodar o comando! <<<

%%writefile overflow.c
#include <stdio.h>
#include <locale.h>

#define RESET   "\033[0m"
#define VERDE   "\033[1;32m"
#define LARANJA "\033[38;5;215m"
#define VERMELHO "\033[1;31m"

int main() {
    setlocale(LC_ALL, "Portuguese");

    printf(VERDE "=== TESTE DE ESTOURO DE MEMÓRIA (OVERFLOW) ===\n\n" RESET);

    short numero = 32767;
    printf("Valor Máximo do Short (2 bytes): " LARANJA "%d\n" RESET, numero);

    numero = numero + 1;
    printf(VERMELHO "Após somar +1 (Overflow): %d\n" RESET, numero);

    return 0;
}
```

*Execute com:*

```python
# Compila o arquivo criado, gera um executável e executa o programa
!gcc overflow.c -o overflow && ./overflow
```
