##**Camada de Aplicação: Resolução DNS**

1. **DNS**: Sintaxe e Resolução de Nomes O DNS traduz nomes legíveis (sintaxe humana) em endereços IP (endereçamento lógico). Abaixo, simulamos essa transição que ocorre na Camada de Aplicação para localizar servidores na rede.

```python
import socket

def validar_dns():
    print(f"{'DOMÍNIO ANALISADO':<35} | {'ENDEREÇO IP':<15}")
    print("-" * 55)

    # Domínios para teste de resolução
    alvos = ["google.com", "github.com", "murilo-guimaraes.github.io"]

    for host in alvos:
        try:
            ip = socket.gethostbyname(host)
            print(f"{host:<35} | {ip:<15}")
        except socket.gaierror:
            print(f"{host:<35} | Falha na resolução")

validar_dns()
```

##**Camada de Transporte e Semântica**

2. **Protocolo HTTP**: Semântica e Metadados
Ao realizar uma requisição, analisamos a semântica da resposta. Os cabeçalhos (headers) enviados pelo servidor confirmam o estado da transação e os protocolos utilizados na Camada de Transporte e Aplicação.

```python
import requests

def inspecionar_http(url):
    print(f"Iniciando requisição para: {url}\n")
    try:
        # A requisição GET valida a interação entre as camadas superiores
        response = requests.get(url, timeout=5)

        print(f"Status da Resposta: {response.status_code}")
        print(f"Servidor: {response.headers.get('Server', 'Não identificado')}")
        print(f"Tipo de Conteúdo: {response.headers.get('Content-Type')}")
        print(f"Data no Servidor: {response.headers.get('Date')}")

    except Exception as e:
        print(f"Erro na conexão: {e}")

inspecionar_http("https://api.github.com")
```
